```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
empresas = ['Apple','Google','Amazon','Microsoft','Tecent','McDonalds','Visa','Facebook','Alibaba','Loouis Vuitton']
bilhoes = [947.062,819.573,705.646,611.460,214.023,196.526,191.032,186.421,169.966,124.273]
```

```python
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhoes')
```

```python
plt.rcParams["figure.figsize"] = (10,3)
plt.plot(empresas, bilhoes)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhoes')
```

```python
#Cor da linha
plt.plot(empresas, bilhoes, color = 'green')
plt.rcParams["figure.figsize"] = (10,3)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas',fontsize=20)
plt.ylabel('Bilhoes',fontsize=20)
```

```python
#Espessura da linha
plt.plot(empresas, bilhoes, color = 'grey',marker='o', lw = 2)
plt.rcParams["figure.figsize"] = (10,3)
plt.title('Empresas mais ricas do mundo')
plt.xlabel('Empresas',fontsize=20)
plt.ylabel('Bilhoes',fontsize=20)
plt.legend(['Bilhes de dlares'],loc = 'right', fontsize=15)
plt.show()

#lista de comandos
#lw => espessura da linha
```

```python
#mudando o tipo do grfico e colocando a legenda acima
plt.bar(empresas, bilhoes, color = 'green')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhoes',fontsize=20)
plt.title('Empresas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhoes de dlares'], loc = 'high right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
#Grfico mais Simples

plt.bar(empresas, bilhoes,color='grey')
```

```python
cores = ['red', 'yellow', 'green', 'blue', 'purple', 'orange', 'magenta', 'white', 'brown', 'black']

plt.bar(empresas, bilhoes, color = cores, edgecolor='black')
plt.title('Empresas mais ricas do mundo',fontsize = 25)
plt.ylabel('Bilhoes de dlares', fontsize=20)
plt.xlabel('Empresas',fontsize=20)
plt.grid(True)
plt.show()
```

```python
#barh - barras horizontais
plt.style.use('seaborn-dark')
plt.barh(empresas,bilhoes, color=cores,edgecolor='black')
plt.title('Empresas mais ricas do mundo',fontsize=25)
plt.ylabel('Bilhoes de dolares', fontsize=20)
plt.xlabel('Empresas',fontsize=20)
plt.grid(True)
```
